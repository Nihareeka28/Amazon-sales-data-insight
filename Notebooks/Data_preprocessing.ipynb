# Data Cleaning

```python
df[df['rating_count'].isnull()]
```

```python
df['rating_count']=df['rating_count'].fillna(0)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.dtypes
```

```python
df['discounted_price']=df['discounted_price'].replace({'₹':'',',':''},regex=True)
df['actual_price']=df['actual_price'].replace({'₹':'',',':''},regex=True)
df['discount_percentage']=df['discount_percentage'].replace({'%':''},regex=True)
df['rating_count']=df['rating_count'].replace({',':''},regex=True)
```

```python
df['discounted_price']=df['discounted_price'].astype(float)
df['actual_price']=df['actual_price'].astype(float)
df['discount_percentage']=df['discount_percentage'].astype(float)
df['rating_count']=df['rating_count'].astype(int)
```

```python
df['rating']=df['rating'].astype(float)
```

```python
df['rating'].str.contains('\|', na=False).any()
```

```python
df['rating'].str.contains('\|', na=False).sum()
```

```python
df[df['rating'].str.contains('\|', na=False)]
```

```python
df['rating'] = df['rating'].str.replace('|', '0', regex=False)
```

```python
df['rating'] = df['rating'].str.strip()
```

```python
df['rating'].replace('', '0', inplace=True)
```

```python
df['rating'] = df['rating'].astype(float)
```

# Feature Engineering

```python
df.head()
```

```python
#price difference
df['price_difference']=df['actual_price']-df['discounted_price']
```

```python
#price ratio-it will define the discount strength
df['discount_ratio']=df['discounted_price']/df['actual_price']
```

```python
#discount_category-for pattern and pricing stratergy

def discount_group(x):
    if x>50:
        return "High discount"
    elif x>20:
        return "Meduim discount"
    else:
        return "low discount"

df['discount_category']=df['discount_percentage'].apply(discount_group)
```

```python
#rating category- for plots and grouping

def rating_group(x):
    if x>4.0:
        return "Excellent"
    elif x>3.0:
        return "Good"
    else:
        return "poor"

df['rating_category']=df['rating'].apply(rating_group)
```

```python
#review_wordcount-this feature will give deeper insights into customer_behaviour

df['review_wordcount'] = df['review_content'].apply(lambda x: len(str(x).split()))
```
